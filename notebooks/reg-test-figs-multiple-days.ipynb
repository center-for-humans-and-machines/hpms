```python
from pathlib import Path

import polars as pl

date_folders = [
    d for d in ratings_dir.iterdir() if d.is_dir() and d.name.startswith("2025-08-")
]
date_folders.sort()

for date_folder in date_folders:
    date = date_folder.name
    print(f"Processing {date}...")

    for round_num in [2, 3]:
        round_key = f"round_{round_num}"
        llama_file = (
            date_folder
            / f"rated-llama-guard-dataset-round-{round_num}-gpt-4o-mini-2024-07-18.json"
        )
        print(f"File: {date_folder / llama_file.name}")

        df = pl.read_json(llama_file.expanduser())

        # Check for messages with multiple categories
        # Split llama_guard_score by comma and count categories per message
        # ...existing code...
        df_with_categories = df.with_columns(
            [
                pl.col("llama_guard_score")
                .str.split(",")
                .list.eval(
                    pl.element().str.strip_chars(" ")
                )  # Apply strip to each element in the list
                .alias("categories_list"),
                pl.col("llama_guard_score")
                .str.split(",")
                .list.len()
                .alias("num_categories"),
            ]
        )
        # ...existing code...

        # Filter for messages with multiple categories (excluding "0" which means safe)
        multiple_categories = df_with_categories.filter(
            (pl.col("num_categories") > 1) & (pl.col("llama_guard_score") != "0")
        )

        print(f"Round {round_num} - Total messages: {len(df)}")
        print(
            f"Round {round_num} - Messages with multiple categories: {len(multiple_categories)}"
        )

        if len(multiple_categories) > 0:
            print("Examples of multiple categories:")
            for row in multiple_categories.head(5).iter_rows(named=True):
                print(
                    f"  Categories: {row['llama_guard_score']} -> {row['categories_list']}"
                )

        # Original single category analysis
        categories_df = (
            df.filter(pl.col("llama_guard_score") != "0")
            .group_by("llama_guard_score")
            .agg(pl.len().alias("count"))
            .sort("llama_guard_score")
        )

        categories_count = dict(
            zip(categories_df["llama_guard_score"], categories_df["count"])
        )
        print(f"Round {round_num} categories count: {categories_count}")
        print("-" * 50)
```
